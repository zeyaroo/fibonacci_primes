```python
#Fibonacci sequence: 0,1,1,2,3,5,8,13,21,.........
#measure execution time
import timeit
#...
import time
import sys
sys.setrecursionlimit(3000)
mem = dict()

def memoized_fib(n):

    if n <= 1:
        return 0
    if n <=2:
        return 1
    if n in mem:
        return mem[n]
    else:
        mem[n] = memoized_fib(n-1)+memoized_fib(n-2)
        return mem[n]
```

```python
def populate(start, chunksize,nth):
    i = start
    while i < nth:
        if i in mem:
            i += chunksize
            pass
        else:
            memoized_fib(i)
            i +=chunksize
```

```python
populate(1,1,100)
time.sleep(1)
populate(100,10,1000)
time.sleep(1)
populate(1000,100,10000)
time.sleep(1)
populate(10000,1000,100000)
time.sleep(1)
```

```python
#Miller Rabin Primality Test
import random
def is_prime_miller_rabin(n, k=5):
    if n <= 1:
        return False
    if n <= 3:
        return True

    # Write n as (2^r) * d + 1
    r, d = 0, n - 1
    while d % 2 == 0:
        r += 1
        d //= 2

    # Witness loop
    for _ in range(k):
        a = random.randint(2, n - 2)
        x = pow(a, d, n)

        if x == 1 or x == n - 1:
            continue

        for _ in range(r - 1):
            x = pow(x, 2, n)
            if x == n - 1:
                break
        else:
            return False  # Definitely composite

    return True  # Likely prime
```

```python
#is the 1000th Fibonacci sequence a prime?
is_prime_miller_rabin(mem[1000])
```

```python
def random_number(start,end):
    return random.randint(start, end)
##PROBABLY a prime
while True:
  rand = random_number(100,500)
  if(is_prime_miller_rabin(mem[rand])):
    print("This {}th Fibonacci sequence is *PROBABLY* a prime number {}".format(rand,mem[rand]))
    break
```

```python
#Quite confident it is a prime
import sympy
while True:
  rand = random_number(500,1000)
  if(is_prime_miller_rabin(mem[rand])):
    if(sympy.isprime(mem[rand])):
      print("This {}th Fibonacci sequence is definitely a prime number {}".format(rand,mem[rand]))
    break
```
